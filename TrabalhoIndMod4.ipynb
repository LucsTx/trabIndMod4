```python
import pandas as pd
import matplotlib.pyplot as plt
```

## <font color = red>  Gerar um relatório de progresso diário para mostrar o quão produtivo está sendo o trabalho dos funcionários

```python
tabDados = {'Dia':['Segunda','Terça','Quarta','Quinta','Sexta','Sábado','Domingo'],'Horas Trabalhadas':[6,7,8,6,7,5,4], 'Bugs Corrigidos':[3,2,1,4,3,2,1],'Tarefas Concluídas':[5,4,6,4,5,3,2]}
tabDados
```

```python
df = pd.DataFrame (tabDados)
df
```

### DISTRIBUIÇÂO DAS HORAS TRABALHADAS

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Dia'], df['Horas Trabalhadas'])
plt.xlabel('Dia')
plt.ylabel('Horas Trabalhadas')
plt.title('Distribuição das Horas Trabalhadas')
plt.show()





```

### BUGS CORRIGIDOS E TAREFAS CONCLUIDAS

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Dia'], df['Bugs Corrigidos'], marker='o', label='Bugs Corrigidos')
plt.plot(df['Dia'], df['Tarefas Concluídas'], marker='o', label='Tarefas Concluídas')
plt.xlabel('Dia')
plt.ylabel('Quantidade')
plt.title('Bugs Corrigidos e Tarefas Concluídas')
plt.legend()
plt.grid()
plt.show()
```

### SOMA TOTAL E MÉDIAS DE HORAS TRABALHADAS, BUGS CORRIGIDOS E TAREFAS CONCLUÍDAS

```python
df['Total de Horas Trabalhadas'] = df['Horas Trabalhadas'].sum()
df['Total de Bugs Corrigidos'] = df['Bugs Corrigidos'].sum()
df['Total de Tarefas Concluídas'] = df['Tarefas Concluídas'].sum()

print(df)
```

#### Calcular a Produtividade Diária (Tarefas Concluídas por Hora)

```python
df['Produtividade Diária'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']

print(df)
```

#### VISUALIZAÇÃO PRODUTIVIDADE DIÁRIA

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Dia'], df['Produtividade Diária'], marker='o')
plt.xlabel('Dia')
plt.ylabel('Produtividade Diária (Tarefas Concluídas por Hora)')
plt.title('Produtividade Diária')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
